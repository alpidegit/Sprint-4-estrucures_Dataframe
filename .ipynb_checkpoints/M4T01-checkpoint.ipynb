Como parece estar bien, procedo a hacer los ejercicios pero ante voy a verificar que, como parece, por cada trimestre tenemos el precio de alquiler medio mensual y por superficie, es decir que "Lloguer_mitja" y "Preu"son columnas que a su vez contienen  dos informaciones distintas:

```python
df.groupby("Nom_Districte")["Lloguer_mitja"].value_counts()
```

Lo lógico sería pasar la columna de Lloguer mitja a dos, una con los precios medio mensuales y otra con los precios por superficie, así que lo haré para tener la df1 con las columnas separadas, aunque utilizaré df cuando pueda:

```python
df_pivot = pd.pivot_table(df, values = "Preu", index = ["Any", "Trimestre", "Codi_Districte", "Nom_Districte", "Codi_Barri", "Nom_Barri"],
                          columns = ["Lloguer_mitja"], aggfunc='first')
df_pivot
```

```python
# Reseteo el índice y renombrar las columnas
df1 = df_pivot.reset_index().rename(columns={"Lloguer mitjà mensual (Euros/mes)": "Lloguer €/mes", "Lloguer mitjà per superfície (Euros/m2 mes)": "Lloguer €/m2 mes"})
df1
```

### - No aporta gran cosa, pero para voy a empezar graficando 1 sola variable: vamos a ver las veces que se evalua el precio por distrito (divido por 2 por tener los precios por mes y por superficie en df). lógicamente, los resultados me salen proporcionales a la cantidad de barrios que tiene en distrito:

```python
counts = df["Nom_Districte"].value_counts()/2   # Divido por dos dado que tener el rpecio por mes y superficie me duplica los registros

fig, (ax1, ax2) = plt.subplots(1,2, figsize = (15,5))

ax1.set_title("Cantidad de medidas por Distrito", pad=20) #pad= define la distancia entre el titulo y el limite superior del grafico
ax1.set_ylabel("Cantidad de medidas")
ax1.bar(counts.index, counts)
ax1.set_xticks(range(len(counts.index)))
ax1.set_xticklabels(counts.index, rotation = 75)

sns.set_theme()                               # me sirve para separar ligeramente todos los quesitos a lav ez de forma sencilla
ax2.pie(counts, labels = counts.index, autopct = "%0.1f %%")
ax2.set_title("Cantidad % de medidas por Distrito", pad=20)

plt.show()
fig.savefig('figure1.png', dpi=300)
```

### - Si hacemos un histograma de distribución de precios se ve claramente que tenemos dos muestras bien diferenciadas:

```python
sns.set_style("whitegrid")
plt.figure(figsize=(8,6))
sns.histplot(data=df, x="Preu", bins=80, kde=True, hue="Lloguer_mitja")
plt.title("Distribución de precios", pad=20)
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
fig.savefig('figure2.png', dpi=300)
```

Es evidente que no tiene sentido poner en la misma gráfico el precio por m^2 (con valores alrededor de 12 € de media) y el precio promedio mesnual de alquiler (con promedios cercanos a los 875€ y máximos alrededor de 1800€).

### - Veamos las graficas de frecuencias de precios separadas:

```python
sns.set_style("whitegrid")
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,8), sharey = True)  #sharey=True para comparir los ejes Y

sns.histplot(data=df1, x="Lloguer €/m2 mes", bins=50, kde=True, ax=ax1)
ax1.set_title("Distribución de precios en €/m^2", pad=20)
ax1.set_xlabel("Precio en €/m^2")
ax1.set_ylabel("Frecuencia")

sns.histplot(data=df1, x="Lloguer €/mes", bins=50, kde=True, ax=ax2)
ax2.set_title("Distribución de precios de alquiler mensual", pad=20)
ax2.set_xlabel("Precio en €/mes")
ax2.set_ylabel("Frecuencia")
ax2.tick_params(axis='y', labelleft=True)

plt.show()
fig.savefig('figure3.png', dpi=300)
```

Observamos los siguiente estadísticos:

```python
desc_m2 = df1["Lloguer €/m2 mes"].describe().round(1)
desc_mes = df1["Lloguer €/mes"].describe().round(1)

desc_df = pd.concat([desc_m2, desc_mes], axis=1)
print(desc_df)
```

Ya nos adelantan los graficos que la diferncia de precios por m^2 alrededor de su media de 12.7 € es bastante parecida a una campana de gaus con +/- k€ de difernencia maxima aproximada, sin embargo el precio por alquiler mensual tiene la moda desplazada a la izquierda con distritos con precios muy elevados de hasta 1800 euros mensuales.

### - Grafiquemos un swarmplot para ver la dsitribución de precios en función del Distrito:

```python
sns.set_style("whitegrid")
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15,10), sharex = True)  #sharey=True para comparir los ejes Y

sns.swarmplot(data=df1, x="Nom_Districte", y="Lloguer €/m2 mes", ax=ax1)
ax1.set_title("Distribución de precios en €/m^2 por Distritos", pad=15)
ax1.set_xlabel("Distritos")
ax1.set_ylabel("Precios en €/m^2")

sns.swarmplot(data=df1, x="Nom_Districte", y="Lloguer €/mes", ax=ax2)
ax2.set_title("Distribución de precios de alquiler mensual por Ditritos", pad=15)
ax2.set_xlabel("Distritos")
ax2.set_ylabel("Precio mensual en €/mes")

plt.show()
fig.savefig('figure4.png', dpi=300)
```

Como vemos en el gráfico, tenemos distritos como Sarrià ciutat vella, les corts y Sant Martí donde el precio por metro cuandrado puede ser mas elevado en alguno de suss barrios, comparado con el resto de Distritos, y sin embargo el precio mensual en Ciutat Vella está en la media, teniendo Les Corts, Sarrià y Sant Marti barrios con precios elevados por m^2.

### - Veámoslo en graficos de cajas tambien para ver las características estadísitcas de precios por cada Distrito:

```python
sns.set_style("whitegrid")
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15,10), sharex = True)  #sharey=True para compartir los ejes Y

sns.boxplot(data=df1, x="Nom_Districte", y="Lloguer €/m2 mes", ax=ax1)
ax1.set_title("Distribución de precios en €/m^2 por Distritos", pad=15)
ax1.set_xlabel("Distritos")
ax1.set_ylabel("Precios en €/m^2")

sns.boxplot(data=df1, x="Nom_Districte", y="Lloguer €/mes", ax=ax2)
ax2.set_title("Distribución de precios de alquiler mensual por Ditritos", pad=15)
ax2.set_xlabel("Distritos")
ax2.set_ylabel("Precio mensual en €/mes")

plt.show()
fig.savefig('figure5.png', dpi=300)
```

A la luz de lso gráficos, puede haber un efecto de la cantidad de pisos disponibles en alquiler o de las dimensiones de los pisos, es decir el total de m^2 disponibles frente al precio de alquiler. Tambien observamos que hay Distritos con grandes difernecias entre barrios, por ejemplo, Sarrià tiene un precio medio por m^2 mas alto que Les Corts en todos sus barrios, sin embargo Les Corts tiene barrios con alquilers mensuales mucho mas altos mientras que la mayor parte son mas baratos. En el casodedel les Corts deberíamos saber si los pisos son de media mas grandes o es un efecto, como digo, solo dependiente de la localización de los barrios.

### - Grafiquemos en un heatmap el precio medio por Distrito y Trimestre:

```python
# introduzco aggfunc="mean" para trabajar con los valores medios de "Llogier €/mes"
df_heatmap = df1.pivot_table(values="Lloguer €/mes", index="Nom_Districte", columns="Trimestre", aggfunc="mean")

fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(df_heatmap, cmap="coolwarm", annot=True, fmt=".0f", cbar_kws={"label": "Precio de alquiler"}, ax=ax)
# cmap='coolwarm': Define la paleta de colores que se utiliza.
# annot=True: Para mostrar los valores numericos en las celdas.
# fmt='.0f': Define el formato: .0f = números enteros sin decimales.
# cbar_kws={'label': 'Precio de alquiler (€/mes)'}: Define la etiqueta de la paleta de colores.
ax.set_xlabel("Trimestre")
ax.set_ylabel("Distrito")
ax.set_title("Precio medio en (€/mes) por distrito y trimestre", pad=20)
plt.show()
fig.savefig('figure12.png', dpi=300)
```

### Hago los grafico de precios por barrios de Barrcelona y los guardo com imagen y como html:

```python
import matplotlib.pyplot as plt                            #Tambien lo he puesto al inicio para recopilar todo pero lo dejo como geopandas antes para saber cunado lo cargo
from mpl_toolkits.axes_grid1 import make_axes_locatable

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(30,20))

data1 = dfb.plot(column="Lloguer €/mes", cmap="plasma", zorder=5, ax=ax1)
ax1.set_title("Alquiler en €/mes por barrios de Barcelona", pad = 20)
ax1.set_xlabel("Longitud")
ax1.set_ylabel("Latitud")
ax1.grid(False)                     # Desactiva la cuadrícula porque daba un error aunque graficaba igual
divider1 = make_axes_locatable(ax1) # Se crea un nuevo eje de la figura ax1 para la barra de colores. 
cax1 = divider1.append_axes("right", size="5%", pad=0.2) #La función append_axes se utiliza para crear un nuevo eje adjunto al eje ax1 en la posición "right" y con un ancho del "5%" y un espacio de "0.2" con respecto a ax1.
cax1.grid(False)                     # Desactiva la cuadrícula del nuevo eje porque daba un error aunque graficaba igual
fig.colorbar(data1.get_children()[0], label="€/mes", cax=cax1)  # Agrega una barra de color a la figura que muestra la relación entre los valores de los datos y los colores utilizados para visualizarlos en ax1.
 
data2 = dfb.plot(column="Lloguer €/m2 mes", cmap="plasma", zorder=5, ax=ax2)
ax2.set_title("Alquiler en €/m^2 por barrios de Barcelona", pad = 20)
ax2.set_xlabel("Longitud")
ax2.set_ylabel("Latitud")
ax2.grid(False)  # Desactiva la cuadrícula
divider2 = make_axes_locatable(ax2)
cax2 = divider2.append_axes("right", size="5%", pad=0.2)
cax2.grid(False)
fig.colorbar(data2.get_children()[0], label="€/m^2", cax=cax2)

plt.show()
fig.savefig('figure13.png', dpi=600)
mpld3.save_html(fig, 'figura13.html')
```
